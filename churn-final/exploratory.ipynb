```python
def prepare_df(df):
    df["Senior"].replace(0, "No", inplace=True)
    df["Senior"].replace(1, "Yes", inplace=True)
    df["TotalCharges"].replace(" ", np.NaN, inplace=True)
    df["TotalCharges"] = pd.to_numeric(df["TotalCharges"])
    df["TotalCharges"].fillna(0.0, inplace=True)
    
prepare_df(train)
```

```python
categorical_cols = list(set(train) - set(numeric_cols))

def countplots(hue=None):
    cols_count = 4
    rows_count = (len(categorical_cols) + cols_count - 1) // cols_count
    fig, ax = plt.subplots(rows_count, cols_count, figsize=(16, 36))

    for i in range(rows_count):
        for j in range(cols_count):
            if cols_count * i + j >= len(categorical_cols):
                break
            cur_ax = sns.countplot(
                x=categorical_cols[cols_count * i + j], 
                data=train, 
                ax=ax[i][j], 
                hue=hue,
            )
            cur_ax.set_xticklabels(
                cur_ax.get_xticklabels(), 
                fontsize=10, 
                rotation=15, 
                ha="right",
            )
```

```python
countplots()
```

```python
countplots("Churn")
```
